Loan Predictor

Loading Dataset 

```python
import pandas as pd 

df = pd.read_csv("../dataset/loan_data.csv")
```

```python
output_column = df["loan_status"]
output_column
```

```python
input_columns = df.drop("loan_status",axis=1)
input_columns
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(input_columns, output_column, test_size=0.25, random_state=42)

```

Model Training

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

lr.fit(X_train, y_train)
```

Predicting Value after training

```python
y_train_predicted_value  = lr.predict(X_train)
y_test_predicted_value = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Calculate metrics for the training data
lr_train_mse = mean_squared_error(y_train, y_train_predicted_value)
lr_train_r2 = r2_score(y_train, y_train_predicted_value)
lr_train_mae = mean_absolute_error(y_train, y_train_predicted_value)

# Calculate metrics for the test data
lr_test_mse = mean_squared_error(y_test, y_test_predicted_value)
lr_test_r2 = r2_score(y_test, y_test_predicted_value)
lr_test_mae = mean_absolute_error(y_test, y_test_predicted_value)

# Print the evaluation metrics
print('LR MSE (Train):', lr_train_mse)
print('LR R2 (Train):', lr_train_r2)
print('LR MAE (Train):', lr_train_mae)

print('LR MSE (Test):', lr_test_mse)
print('LR R2 (Test):', lr_test_r2)
print('LR MAE (Test):', lr_test_mae)
```

```python
results = pd.DataFrame(["Linear Regression" , lr_train_mse , lr_train_r2 ,lr_train_mae, lr_test_mse , lr_test_r2 , lr_test_mae]).transpose()
results.columns = ['Method', 'Training MSE', 'Training R2',"Training MAE", 'Test MSE', 'Test R2' , "Test MAE"]

results
```

Naive Bayes training

```python
from sklearn.naive_bayes import GaussianNB

# Initialize the Gaussian Naive Bayes model
nb = GaussianNB()

# Train the model
nb.fit(X_train, y_train)

y_nb_train_predicted_value = nb.predict(X_train)
y_nb_test_predicted_value = nb.predict(X_test)


# Calculate metrics for the training data
nb_train_mse = mean_squared_error(y_train, y_nb_train_predicted_value)
nb_train_r2 = r2_score(y_train, y_nb_train_predicted_value)
nb_train_mae = mean_absolute_error(y_train, y_nb_train_predicted_value)

# Calculate metrics for the test data
nb_test_mse = mean_squared_error(y_test, y_nb_test_predicted_value)
nb_test_r2 = r2_score(y_test, y_nb_test_predicted_value)
nb_test_mae = mean_absolute_error(y_test, y_nb_test_predicted_value)

# Print the evaluation metrics
print('NB MSE (Train):', nb_train_mse)
print('NB R2 (Train):', nb_train_r2)
print('NB MAE (Train):', nb_train_mae)

print('NB MSE (Test):', nb_test_mse)
print('NB R2 (Test):', nb_test_r2)
print('NB MAE (Test):', nb_test_mae)


nb_results = pd.DataFrame(["Naive Bayes" , nb_train_mse , nb_train_r2 ,nb_train_mae, nb_test_mse , nb_test_r2 , nb_test_mae]).transpose()
nb_results.columns = ['Method', 'Training MSE', 'Training R2',"Training MAE", 'Test MSE', 'Test R2' , "Test MAE"]

df_models = pd.concat([results, nb_results], axis=0)
df_models
```
